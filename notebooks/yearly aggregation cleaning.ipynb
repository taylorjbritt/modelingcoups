```python
import pandas as pd
```

```python
year_agg = pd.read_pickle('../data/year_agg.pkl')
```

```python
nmc_df = pd.read_csv('../data/NMC_5_0.csv')
trade_df = pd.read_csv('../data/COW_trade_3.0/national_trade_3.0.csv')
```

```python
nmc_df['yearcode'] = nmc_df['ccode']*10000 + nmc_df['year']
trade_df['yearcode'] = trade_df['ccode']*10000 + trade_df['year']
```

```python
nmc_df
```

```python
year_agg
```

```python
joint_df = year_agg.join(trade_df.set_index('yearcode'), on='yearcode', how = 'inner', rsuffix = '_tradedf' )
df = joint_df.join(nmc_df.set_index('yearcode'), on='yearcode', how = 'inner', rsuffix = '_nmcdf' )
```

```python
df
```

```python
df['pt_attempt'] = df['coupyear']
```

```python
df['pt_suc'] = df['coupsuc']
```

```python
df = df.drop(['coupyear', 'coupsuc'], axis =1)
```

```python
df.to_pickle('../data/yearly_nmc_trade.pkl')
```

```python
df
```

```python
dummies = pd.get_dummies(df['government'])
df_dumb = df.join(dummies)
drop_list = ['yearcode', 'country_tradedf', 'year_tradedf', 'alt_imports', 'alt_exports', 'source1', 'source2',
       'version', 'stateabb', 'ccode_nmcdf', 'year_nmcdf', 'version_nmcdf', 'tpop', 'ccode_tradedf', 'Parliamentary Democracy']
df = df_dumb.drop(drop_list, axis = 1)
```
