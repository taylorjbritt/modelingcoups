```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/joint_df.pkl')
```

```python
df.columns
```

```python
dummies = pd.get_dummies(df['government'])
df_dumb = df.join(dummies)
df_dumb['pt_attempt'] = df_dumb['coupyear']
df_dumb['pt_suc'] = df_dumb['coupsuc']
drop_list = ['yearcode', 'Parliamentary Democracy', 'ccode', 'country', 'leader', 'coupyear', 'coupsuc', 'government']
df_drop = df_dumb.drop(drop_list, axis = 1)
```

```python
df_drop.columns
```

```python
df_i = df_drop.interpolate(axis = 0)
```
