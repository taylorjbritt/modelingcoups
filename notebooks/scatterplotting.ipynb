```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/pickles/model_ready_df.pkl')
```

```python
df.columns
```

```python
df_coup = df[df['pt_attempt'] == True]
```

```python
df_no_coup = df[df['pt_attempt'] == False]
```

```python
X1 = df_coup['GDP growth (annual %)']
```

```python
X2 = df_no_coup['GDP growth (annual %)']
```

```python
Y1 = df_coup['Population growth (annual %)']
```

```python
Y2 = df_no_coup['Population growth (annual %)']
```

```python
fig, ax = plt.subplots(figsize = (12, 12))
ax.scatter(X2,Y2,color='blue', alpha = .5)
ax.scatter(X1,Y1,color='red')
ax.set_ylabel('Population Growth')
ax.set_xlabel('GDP Growth')

```

```python
def make_scatter(x_var, y_var):
    X1 = df_coup[x_var]
    X2 = df_no_coup[x_var]
    Y1 = df_coup[y_var]
    Y2 = df_no_coup[y_var]
    fig, ax = plt.subplots(figsize = (16, 16))
    ax.scatter(X2,Y2,color='blue', alpha = .5)
    ax.scatter(X1,Y1,color='red', marker = "^")
    ax.set_ylabel(y_var, fontsize = 14)
    ax.set_xlabel(x_var, fontsize = 14)
    title = x_var + ' vs. ' + y_var
    ax.set_title(title, fontsize = 18)

    
```

```python
df.columns
```

```python
make_scatter('Population ages 0-14 (% of total population)', 'Life expectancy at birth, female (years)')
```

```python
def make_scatter3d(x_var, y_var, z_var, df):
    df_coup = df[df['pt_attempt'] == True]
    df_no_coup = df[df['pt_attempt'] == False]
    X1 = df_coup[x_var]
    X2 = df_no_coup[x_var]
    Y1 = df_coup[y_var]
    Y2 = df_no_coup[y_var]
    Z1 = df_coup[z_var]
    Z2 = df_no_coup[z_var]
    fig, ax = plt.subplots(figsize = (16, 16))
    ax = plt.axes(projection='3d')
    ax.scatter3D(X2,Y2,Z2, cmap='Greens', alpha = .2, label = 'No Coup')
    ax.scatter3D(X1,Y1,Z1, color='red', marker = "^", label = 'Coup')
    ax.set_ylabel(y_var, fontsize = 14)
    ax.set_xlabel(x_var, fontsize = 14)
    ax.set_zlabel(z_var, fontsize = 14)
    ax.legend(loc = 'lower left', fontsize = 20)
    title = x_var + ' vs. ' + y_var + ' vs ' + z_var
    savename = '../images/' + x_var + '_' + y_var + '_' + z_var
    ax.set_title(title, fontsize = 18)
    fig.savefig(savename, dpi=400)
```

```python
from mpl_toolkits import mplot3d
```

```python
make_scatter3d('Population ages 0-14 (% of total population)', 'Life expectancy at birth, female (years)', 'GDP growth (annual %)', df)
```
